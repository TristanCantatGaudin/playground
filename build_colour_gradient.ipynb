```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib as mpl
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.colors as mcolors


# Make some illustrative fake data:
x = np.arange(0, np.pi, 0.1)
y = np.arange(0, 2 * np.pi, 0.1)
X, Y = np.meshgrid(x, y)
Z = np.cos(X) * np.sin(Y) * 10

color_lists = """#d13042
#f95714
#fcb103
#759c36
#1f6c91
#4f4392

#f3f3f4
#c6d624
#e56cc0
#472e97
#281367
#151517

#e928c8
#85a3b2
#732553
#ff5c8d
#1e3442
#142030

#f9f3dd
#f2e7d1
#3ffd6a
#591b98
#442266
#110320

#f1f0ef
#dfdde0
#66eebb
#8d5eb7
#57317e
#151632

#f7fcff
#88eeee
#ff2077
#bb1166
#380282
#100438

#c8f9cc
#80ee98
#57cc98
#37a3a5
#225779

#d1f4de
#e4c1f9
#a8def8
#ff99c8
#fcf6bc""".split('\n\n')

for color_list in color_lists:

    color_list = color_list.split('\n')
    color_list_rgba = [mcolors.to_rgba(color) for color in color_list]

    n_bins = [len(color_list), 100]
    cmap_name = 'custom_cmap'

    fig, axs = plt.subplots(1, 2, figsize=(8, 3))
    for n_bin, ax in zip(n_bins, axs.flat):
        # Create the colormap
        cmap = LinearSegmentedColormap.from_list(cmap_name, color_list_rgba, N=n_bin)
        # Fewer bins will result in "coarser" colomap interpolation
        im = ax.imshow(Z, origin='lower', cmap=cmap)
        ax.set_title("N bins: %s" % n_bin)
        fig.colorbar(im, ax=ax)
    plt.tight_layout()
    plt.show()
```
